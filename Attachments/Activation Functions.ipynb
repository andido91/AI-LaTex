# Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (7.2, 4.4),
                     'axes.grid':      True,
                     'grid.color':     'gainsboro',
                     'font.size':      13})
```

## Sigmoid
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle\sigma(x)=\frac{1}{1+e^{-x}}=\frac{e^x}{e^x+1}$

```python
def Sigmoid(x):
    return 1/(1 + np.exp(-x))
def Derivative(x):
    return np.exp(x)/(np.exp(x) + 1)**2
```

```python
x = np.linspace(-4, 4, 201)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, Sigmoid(x), 'k-')
plt.title('Sigmoid')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_Sigmoid.pdf')
plt.show()
```

## Hyperbolic tangent
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle f(x)=\tanh{x}$

```python
def Tanh(x):
    return np.tanh(x)
def Derivative(x):
    return 1/(np.cosh(x))**2
```

```python
x = np.linspace(-4, 4, 201)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, Tanh(x), 'k-')
plt.title('Hyperbolic tangent')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_Hyperbolic_Tangent.pdf')
plt.show()
```

## ReLU
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle f(x)=x\Theta(x)=\max{\{0,x\}}=\begin{cases}
x\,,&\text{if $x\geq0$}\\
0\,,&\text{if $x<0$}
\end{cases}$

```python
def ReLU(x):
    return x*np.heaviside(x, 0)
def Derivative(x):
    return np.piecewise(x, [x < 0, x > 0], [0, 0])
```

```python
x = np.linspace(-4, 4, 201)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, ReLU(x), 'k-')
plt.title('ReLU')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_ReLU.pdf')
plt.show()
```

## Leaky ReLU
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle f(x)=\begin{cases}
x\,,&\text{if $x\geq0$}\\
0.01x\,,&\text{if $x<0$}
\end{cases}$

```python
def Leaky_ReLU(x):
    return np.piecewise(x, [x < 0, x >= 0], [lambda x: 0.01*x, lambda x: x])
def Derivative(x):
    return np.piecewise(x, [x < 0, x > 0], [0.01, 0])
```

```python
x = np.linspace(-1, 1, 501)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, Leaky_ReLU(x), 'k-')
plt.title('Leaky ReLU')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_Leaky_ReLU.pdf')
plt.show()
```

## Swish or SiLU
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle f(x)=x\sigma(x)=\frac{x}{1+e^{-x}}$

```python
def SiLU(x):
    return x/(1 + np.exp(-x))
def Derivative(x):
    return np.exp(x)*(x + np.exp(x) + 1)/(np.exp(x) + 1)**2
```

```python
x = np.linspace(-4, 4, 201)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, SiLU(x), 'k-')
plt.title('Swish or SiLU')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_SiLU.pdf')
plt.show()
```

## Mish
&nbsp;&nbsp;&nbsp;&nbsp;
$\displaystyle f(x)=x\tanh{\bigl[\ln{(1+e^x)}\bigr]}$

```python
def Mish(x):
    return x*np.tanh(np.log(1+np.exp(x)))
def Derivative(x):
    return np.exp(x)*(4*(x + 1) + 4*np.exp(2*x) + np.exp(3*x) + np.exp(x)*(4*x + 6))/(2*np.exp(x) + np.exp(2*x) + 2)**2
```

```python
x = np.linspace(-4, 4, 201)

plt.plot(x, Derivative(x), 'r-')
plt.plot(x, Mish(x), 'k-')
plt.title('Mish')

#plt.savefig('C:/Users/Fre/Desktop/Scuola, Università, Conservatorio/Università/Magistrale/Artificial Intelligence (Drago)/Appunti/Immagini/Activation_Function_Mish.pdf')
plt.show()
```
